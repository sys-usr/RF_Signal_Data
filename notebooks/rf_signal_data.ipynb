```python
#Imports
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import ast
import mplcursors
import mpldatacursor
```

```python


signals = pd.read_csv('../data/logged_data.csv')

display(signals.describe())

display(signals.info())

display(signals.head(10))

display(signals.shape)
```

```python
signals = signals.drop('I/Q Data', axis=1)
```

```python
display(signals.describe())

display(signals.info())

display(signals.head(10))

display(signals.shape)
```

```python
columns_to_analyze = ['Timestamp','Frequency','Signal Strength','Modulation','Bandwidth','Location','Device Type','Antenna Type','Temperature','Humidity','Wind Speed','Precipitation','Weather Condition','Interference Type','Battery Level','Power Source','CPU Usage','Memory Usage','WiFi Strength','Disk Usage','System Load','Latitude','Longitude','Altitude(m)','Air Pressure','Device Status']

# Compute the correlation matrix
corr_matrix = signals[columns_to_analyze].corr()

# Iterate over the correlation matrix and print the correlated categories
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            correlation = corr_matrix.loc[col1, col2]
            if correlation >= 0.5 or correlation <= -0.5:
                print(f"Correlation between {col1} and {col2}: {correlation}")
```

```python
column_list = signals.columns.tolist()

# Print the column names
for column in column_list:
    print(column)
```

```python
columns_to_drop = ['Location', 'Battery Level', 'Humidity', 'Wind Speed', 'Power Source', 'CPU Usage', 'Memory Usage', 'Disk Usage', 'System Load', 'Latitude', 'Longitude', 'Altitude(m)', 'Air Pressure']
signals = signals.drop(columns=columns_to_drop)
```

```python
display(signals)
```

```python
unique_frequencies = signals["Frequency"].unique()

display(unique_frequencies)

output_directory = "Data/Frequency_data"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

combined_data = pd.DataFrame()  # Initialize an empty DataFrame for concatenation

for frequency in unique_frequencies:
    frequency_mhz = frequency / 1000000  # Convert frequency to MHz
    frequency_mhz_str = str(int(frequency_mhz)) + "MHz"
    frequency_df = signals[signals["Frequency"] == frequency]
    output_file = os.path.join(output_directory, f"{frequency_mhz_str}.csv")
    frequency_df.to_csv(output_file, index=False)
    
    # Concatenate the individual data frames
    combined_data = pd.concat([combined_data, frequency_df])

# Save the combined data frame as "logged_data_cleaned.csv"
output_file_combined = "Data/logged_data_cleaned.csv"
combined_data.to_csv(output_file_combined, index=False)
```

```python
# Read the cleaned data file
file_path = "../Data/logged_data_cleaned.csv"
data = pd.read_csv(file_path)

# Extract the 'I/Q data' column
iq_data = data['I/Q Data']

# Convert the 'I/Q data' column to a list of complex numbers
iq_data_complex = [complex(num) for num in ast.literal_eval(iq_data[0])]

# Complex Plane Plot
real_part = np.real(iq_data_complex)
imaginary_part = np.imag(iq_data_complex)

plt.scatter(real_part, imaginary_part, marker='o')
plt.xlabel('Real Part (I)')
plt.ylabel('Imaginary Part (Q)')
plt.title('Complex Plane Plot')
plt.grid(True)
plt.show()
```

```python
# Time-Domain Plot
time = np.arange(len(iq_data_complex))  # Assuming constant time interval
real_part = [np.real(num) for num in iq_data_complex]
imaginary_part = [np.imag(num) for num in iq_data_complex]

plt.plot(time, real_part, label='Real (I)')
plt.plot(time, imaginary_part, label='Imaginary (Q)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time-Domain Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Frequency-Domain Plot
frequency = np.fft.fftfreq(len(iq_data_complex))  # Frequency values
fft_result = np.fft.fft(iq_data_complex)  # Perform Fourier Transform
magnitude_spectrum = np.abs(fft_result)

plt.plot(frequency, magnitude_spectrum)
plt.xlabel('Frequency')
plt.ylabel('Magnitude Spectrum')
plt.title('Frequency-Domain Plot')
plt.grid(True)
plt.show()
```

```python
# Constellation Diagram
plt.scatter(iq_data_complex.real, iq_data_complex.imag, marker='o')
plt.xlabel('Real Part (I)')
plt.ylabel('Imaginary Part (Q)')
plt.title('Constellation Diagram')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast
import mplcursors

# Read the cleaned data file
file_path = "../Data/logged_data_cleaned.csv"
data = pd.read_csv(file_path)

# Extract the relevant columns
iq_data = data['I/Q Data']
frequency = data['Frequency']
modulation = data['Modulation']
signal_strength = data['Signal Strength']

# Convert the 'I/Q data' column to a NumPy array of complex numbers
iq_data_complex = np.array([complex(num) for num in ast.literal_eval(iq_data[0])])

# Complex Plane Plot
real_part = np.real(iq_data_complex)
imaginary_part = np.imag(iq_data_complex)

fig, ax = plt.subplots()
scatter = ax.scatter(real_part, imaginary_part, marker='o')

plt.xlabel('Real Part (I)')
plt.ylabel('Imaginary Part (Q)')
plt.title('Complex Plane Plot')
plt.grid(True)

# Create the hover tooltip with custom information
cursor = mplcursors.cursor(scatter)
@cursor.connect("add")
def on_add(sel):
    index = sel.target.index
    sel.annotation.set_text(
        f"Frequency: {frequency[index]}\n"
        f"Modulation: {modulation[index]}\n"
        f"Signal Strength: {signal_strength[index]}"
    )

plt.show()
```
